```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
scatter plot
bar plot
line plot
countplot
Histogram plot
Heatmap plot
```

Line plot: A line plot shows the relationship between two variables on a continuous scale. It is useful for visualizing trends over time or other ordered categories.

Bar plot: A bar plot represents data as a series of bars, with the height of each bar representing the frequency or count of data in a particular category. It is commonly used for comparing the values of different categories.

Histogram: A histogram is used to visualize the distribution of a single variable by dividing it into a set of bins and displaying the frequency of observations within each bin.

Scatter plot: A scatter plot shows the relationship between two variables by displaying the data points as a collection of dots. It is useful for identifying trends, clusters, and outliers in the data.

Heatmap: A heatmap is a graphical representation of data that uses color-coding to represent different values. It is useful for visualizing the correlation between multiple variables or the distribution of data in a two-dimensional space.

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
data =sns.load_dataset('fmri')
```

```python
data.head()
```

```python
sns.lineplot(x= data.timepoint , y = data.signal)
```

```python
data.head()
```

```python
sns.lineplot(x= data.timepoint , y = data.signal , style = 'event', hue = 'region', data= data)
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
ship =sns.load_dataset('titanic')
```

```python
sns.boxenplot( x= ship.pclass, y= ship.age)
```

```python
sns.boxenplot( x= ship.pclass, y= ship.fare)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
daimonds_data = sns.load_dataset('diamonds')
```

```python
daimonds_data.head()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
data_flights = sns.load_dataset('flights')
```
